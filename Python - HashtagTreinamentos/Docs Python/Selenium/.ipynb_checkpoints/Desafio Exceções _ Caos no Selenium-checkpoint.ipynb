## Entrando site jogos do dia

```python
from selenium import webdriver
import time
import pandas as pd



link = "https://pbdatatrader.com.br/jogosdodia"
driver = webdriver.Chrome()

driver.get(link)
```

### Esperando a pagina carregar e entrando nos iframes

```python
while len(driver.find_elements_by_tag_name("iframe")) == 0:
    time.sleep(1)


# entrando primeiro iframe
meu_iframe = driver.find_element_by_tag_name("iframe")
driver.switch_to.frame(meu_iframe)

# entrando segundo iframe
meu_iframe = driver.find_element_by_tag_name("iframe")
driver.switch_to.frame(meu_iframe)



texto = driver.find_element_by_xpath('//*[@id="pvExplorationHost"]/div/div/exploration/div/explore-canvas/div/div[2]/div/div[2]/div[2]/visual-container-repeat/visual-container[19]/transform/div/div[3]/div/visual-modern/div/div/div[2]/div[1]/div[2]/div/div[1]/div').text
print(texto)
```

```python
tabela = driver.find_element_by_class_name("innerContainer")

texto = tabela.text

print(texto)
```

```python
lista_texto = texto.split("\n")

lista_colunas = lista_texto[:11]
lista_valores = lista_texto[11:]

```

```python


dic = {}

for coluna in lista_colunas:
    dic[coluna] = []
    for i in range(20):
        dic[coluna].append(lista_valores[i])
    lista_valores = lista_valores[20:]
    
print(dic)    
```

```python
tabela_df = pd.DataFrame.from_dict(dic)
display(tabela_df)
```

### Implementar o scroll para pegar a tabela inteira

```python
# Código para descobrir qual div tem scroll

# var elements = document.getElementsByTagName("div");

#for (var i = 0; i < elements.length; i++) { if (elements[i].scrollHeight > document.getElementsByTagName("body")[0].scrollHeight) {
#console.log(elements[i]);
#console.log(elements[i].scrollHeight);
#}}
```

```python
for i in range(200):
    posicao_scroll = 250 * i
    driver.execute_script(f"document.getElementsByClassName('bodyCells')[0].scroll(0, {posicao_scroll})")

    # o que eu quero que acontece a cada scroll
    # ler a tabela da pagina
    tabela = driver.find_element_by_class_name("innerContainer")
    texto = tabela.text
    # separar valores e colunas
    lista_texto = texto.split("\n")
    lista_colunas = lista_texto[:11]
    lista_valores = lista_texto[11:]
    
    # criar um dicionario auxiliar
    dic = {}

    for coluna in lista_colunas:
        dic[coluna] = []
        for i in range(20):
            dic[coluna].append(lista_valores[i])
        lista_valores = lista_valores[20:]
     # criar a tabela python temporaria   
    tabela_temporaria = pd.DataFrame.from_dict(dic)
    tabela_df = tabela_df.append(tabela_temporaria, ignore_index = True)
```

```python
# remover duplicatas
tabela_df = tabela_df.drop_duplicates()

# exibir a tabela / armazenar a tabela em excel

display(tabela_df)
```

```python
tabela_df.to_excel("Tabela Dados Jogos.xlsx", index=False)
```
